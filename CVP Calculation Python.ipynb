```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Production Level (Units)": [0, 100, 200, 300, 400, 500],
    "Total Revenue ($)": [0, 5000, 10000, 15000, 20000, 25000],
    "Variable Costs ($)": [0, 2000, 4000, 6000, 8000, 10000],
    "Fixed Costs ($)": [1000, 1000, 1000, 1000, 1000, 1000]
}

df = pd.DataFrame(data)

df.head()
```

```python
# Contribution Margin and Profit
df["Contribution Margin ($)"] = df["Total Revenue ($)"] - df["Variable Costs ($)"]
 
df["Profit ($)"] = df["Contribution Margin ($)"] - df["Fixed Costs ($)"]

print(df[['Contribution Margin ($)','Profit ($)']])
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df["Production Level (Units)"], df["Variable Costs ($)"], label="Variable Costs", marker='o')
plt.plot(df["Production Level (Units)"], df["Profit ($)"], label="Profit", marker='o')

plt.title("Cost-Volume-Profit Analysis")
plt.xlabel("Production Level (Units)")
plt.ylabel("Cost/Profit ($)")
plt.legend()
plt.grid(True)

plt.show()
```
